```python
import copy
import random
import gymnasium as gym
import numpy as np
from datetime import datetime, timedelta
import pickle
import pandas as pd
import time
import imageio.v2 as imageio
```

```python
class Config():
    def __init__(self):
        self.env_name = "Taxi-v4"
        self.total_episode = 1000000
        self.gamma = 0.99
        self.num_action = 6
        self.state_dim = 500
        self.epsilon=1.0
        self.decay_step=int(0.9 * self.total_episode)
        self.eps_min=0.05
        self.test_frequency = 100
        self.seed = 1
        self.test_seeds = [
    1001, 1003, 1007, 1009, 1013, 1019, 1021, 1031, 1033, 1039,
    1049, 1051, 1061, 1063, 1069, 1087, 1091, 1093, 1097, 1103,
    1109, 1117, 1123, 1129, 1151, 1153, 1163, 1171, 1181, 1187,
    1193, 1201, 1213, 1217, 1223, 1229, 1231, 1237, 1249, 1259,
    1277, 1279, 1283, 1289, 1291, 1297, 1301, 1303, 1307, 1319,
    1321, 1327, 1361, 1367, 1373, 1381, 1399, 1409, 1423, 1427,
    1429, 1433, 1439, 1447, 1451, 1453, 1459, 1471, 1481, 1483,
    1487, 1489, 1493, 1499, 1511, 1523, 1531, 1543, 1549, 1553,
    1559, 1567, 1571, 1579, 1583, 1597, 1601, 1607, 1609, 1613,
    1619, 1621, 1627, 1637, 1657, 1663, 1667, 1669, 1693, 1697
]
        #[1, 2, 3, 5, 7, 11, 13, 17, 19, 20]
        self.save_file = "weights/MC-Taxi-v4-1M-run1"

config = Config()
```

```python
random.seed(config.seed)
np.random.seed(config.seed)
```

```python
class Agent():
    def __init__(self, config):
        self.total_episode = config.total_episode
        self.gamma = config.gamma
        self.num_action = config.num_action
        self.state_dim = config.state_dim
        self.epsilon = config.epsilon
        self.decay_step = config.decay_step
        self.eps_min = config.eps_min
        self.decay_per_step = (self.epsilon - self.eps_min) / self.decay_step
        self.test_frequency = config.test_frequency
        self.seed = config.seed
        self.test_seeds = config.test_seeds
        self.save_file = config.save_file
        self.env_name = config.env_name

        self.test_env = gym.make(config.env_name, is_rainy=False, fickle_passenger=False)
        self.env = gym.make(config.env_name, is_rainy=False, fickle_passenger=False)

        self.Q_Table = np.zeros((self.state_dim, self.num_action))
        self.return_counts = np.zeros((self.state_dim, self.num_action))

        self.episode_rewards = []
        self.test_episode_rewards = []
        self.episode_steps = []
        self.test_episode_steps = []
        
    def select_action(self, state):
        if np.random.rand() < self.epsilon:
            return np.random.randint(self.num_action)
        else:
            return np.argmax(self.Q_Table[state])
        
    def decay_epsilon(self):
        self.epsilon = max(self.epsilon - self.decay_per_step, self.eps_min)

    def add(self, state, action = None, episode_return = 0):
        if action is None:
            for a in range(self.num_action):
                self.return_counts[state][a] += 1
                N = self.return_counts[state][a]
                self.Q_Table[state][a] *= (N - 1) / N
                self.Q_Table[state][a] += episode_return / N
        else:
            self.return_counts[state][action] += 1
            N = self.return_counts[state][action]
            self.Q_Table[state][action] *= (N - 1) / N
            self.Q_Table[state][action] += episode_return / N

    def test(self):
        test_episode_rewards = []
        test_episode_steps = []
        
        for test_seed in self.test_seeds:
            state = self.test_env.reset(seed=test_seed)[0]
            self.test_env.action_space.seed(test_seed)
            done = False
            episode_reward = 0.
            episode_step = 0

            while not done:
                action = np.argmax(self.Q_Table[state])
                next_state, reward, terminated, truncated, info = self.test_env.step(action)
                episode_step += 1
                episode_reward += reward
                done = terminated or truncated
                state = next_state
            
            test_episode_steps.append(episode_step)
            test_episode_rewards.append(episode_reward)
        return test_episode_rewards, test_episode_steps
    
    def render(self, seed):
        with open(f"{self.save_file}.pickle", "rb") as f:
            self.Q_Table = pickle.load(f)

        self.render_env = gym.make(self.env_name, is_rainy=False, fickle_passenger=False, render_mode='human')
        state = self.render_env.reset(seed=seed)[0]
        self.render_env.action_space.seed(seed)
        done = False
        episode_reward = 0.

        while not done:
            time.sleep(0.1)
            action = np.argmax(self.Q_Table[state])
            next_state, reward, terminated, truncated, info = self.render_env.step(action)
            episode_reward += reward
            done = terminated or truncated
            state = next_state
        self.render_env.close()
        print(episode_reward)

    def save_gif(self, seed):
        with open(f"{self.save_file}.pickle", "rb") as f:
            self.Q_Table = pickle.load(f)

        self.render_env = gym.make(self.env_name, is_rainy=False, fickle_passenger=False, render_mode="rgb_array")
        state = self.render_env.reset(seed=seed)[0]
        self.render_env.action_space.seed(seed)
        done = False
        episode_reward = 0.
        frames = [self.render_env.render()]

        while not done:
            action = np.argmax(self.Q_Table[state])
            next_state, reward, terminated, truncated, info = self.render_env.step(action)
            frames.append(self.render_env.render())
            episode_reward += reward
            done = terminated or truncated
            state = next_state
        self.render_env.close()
        print(episode_reward)

        imageio.mimsave(
            f"{self.save_file}.gif",
            frames,
            fps=60,
            loop=0
        )

    def learn(self):
        start_time = datetime.now()
        max_test_reward = -10000.

        state = self.env.reset(seed=config.seed)[0]
        self.env.action_space.seed(config.seed)

        self.current_step = 0

        for episode in range(1, self.total_episode+1):
            start_episode_time = datetime.now()
            done = False

            episode_reward = 0.
            epsiode_step = 0
            rewards = []
            states = []
            actions = []

            while not done:
                self.current_step += 1
                action = self.select_action(state)
                next_state, reward, terminated, truncated, info = self.env.step(action)
                episode_reward += reward
                epsiode_step += 1
                rewards.append(reward)
                states.append(state)
                actions.append(action)
                done = terminated or truncated
                state = next_state
            self.episode_steps.append(epsiode_step)
            self.episode_rewards.append(episode_reward)
            self.decay_epsilon()

            self.add(state, None, 0)
            episode_return = 0
            for state, action, reward in zip(states[::-1], actions[::-1], rewards[::-1]):
                episode_return = episode_return * self.gamma + reward
                self.add(state, action, episode_return)
            
            if episode % 1000 == 0:
                print(f"Episode: {episode} Current_Step: {self.current_step} Rewards: {episode_reward} Mean_Rewards: \
{np.array(self.episode_rewards[-min(100, len(self.episode_rewards)):]).mean():.2f}±\
{np.array(self.episode_rewards[-min(100, len(self.episode_rewards)):]).std():.2f} \
Min: {np.array(self.episode_rewards[-min(100, len(self.episode_rewards)):]).min():.2f} \
Max: {np.array(self.episode_rewards[-min(100, len(self.episode_rewards)):]).max():.2f} \
Epsilon: {self.epsilon:.2f} Time: {datetime.now() - start_episode_time} Total_Time: {datetime.now() - start_time}")

            if episode % self.test_frequency == 0:
                start_test_time = datetime.now()
                test_episode_rewards, test_episode_steps = self.test()
                self.test_episode_rewards.append(test_episode_rewards)
                self.test_episode_steps.append(test_episode_steps)

                if max_test_reward < np.array(test_episode_rewards).mean():
                    max_test_reward = np.array(test_episode_rewards).mean()
                    with open(f"{self.save_file}.pickle", "wb") as f:
                        pickle.dump(self.Q_Table, f)
                
                if len(self.test_episode_rewards) % 100 == 0:
                    print(f"Test Episode: {len(self.test_episode_rewards)} Current_Step: {self.current_step} \
Rewards: {np.array(test_episode_rewards).mean():.2f}±{np.array(test_episode_rewards).std():.2f} \
Min: {np.array(test_episode_rewards).min():.2f} Max: {np.array(test_episode_rewards).max():.2f} \
Mean_Rewards: {np.array(self.test_episode_rewards[-min(100, len(self.test_episode_rewards)):]).mean():.2f}±\
{np.array(self.test_episode_rewards[-min(100, len(self.test_episode_rewards)):]).std():.2f} \
Min: {np.array(self.test_episode_rewards[-min(100, len(self.test_episode_rewards)):]).min():.2f} \
Max: {np.array(self.test_episode_rewards[-min(100, len(self.test_episode_rewards)):]).max():.2f} Max_Rewards: {max_test_reward} \
Time: {datetime.now() - start_test_time} Total_Time: {datetime.now() - start_time}")
                    
            state = self.env.reset()[0]

        with open(f"{self.save_file}-log.pickle", "wb") as f:
            pickle.dump({"episode_rewards": self.episode_rewards,
                         "test_episode_rewards": self.test_episode_rewards,
                         "episode_steps": self.episode_steps,
                         "test_episode_steps": self.test_episode_steps,
                         "step": self.current_step}, f)
```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

log_file = "weights/MC-Taxi-v4-1M-run1-log.pickle"
window = 100
test_frequency = config.test_frequency   

with open(log_file, "rb") as f:
    log = pickle.load(f)

episode_rewards = np.array(log["episode_rewards"])
test_episode_rewards = np.array(log["test_episode_rewards"])
episode_steps = np.array(log["episode_steps"])
test_episode_steps = np.array(log["test_episode_steps"])

sum_episode_rewards = np.convolve(episode_rewards, np.ones((window,), dtype=np.float32))[:len(episode_rewards)]
mean_episode_rewards = sum_episode_rewards / np.array([i+1 if i < window else window for i in range(len(episode_rewards))])

test_episode_rewards = test_episode_rewards.mean(-1)
sum_test_episode_rewards = np.convolve(test_episode_rewards, np.ones((window,), dtype=np.float32))[:len(test_episode_rewards)]
mean_test_episode_rewards = sum_test_episode_rewards / np.array([i+1 if i < window else window for i in range(len(test_episode_rewards))])

train_episode = np.arange(1, len(episode_rewards) + 1)
test_episode = np.linspace(
    test_frequency,
    len(episode_rewards),
    len(test_episode_rewards),
    dtype=int,
)

plt.figure(figsize=(10, 6))

plt.plot(
    train_episode,
    mean_episode_rewards,
    label=f"Train (Last-{window}-Average)",
)

plt.plot(
    test_episode,
    mean_test_episode_rewards,
    linewidth=2,
    label=f"Test (Last-{window}-Average)",
)

plt.xlabel("Episode")
plt.ylabel("Reward")
plt.title("Train/Test Reward vs Episode")
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
